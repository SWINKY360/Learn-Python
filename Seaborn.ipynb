```python
#Setup up the notebook
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load data
name = '/pays.csv'
costs = pd.read_csv(name)
```

```python
costs.head()
```

```python
plt.figure(figsize=(17,8))
plt.title("Retirements Costs adjusted for inflation")
sns.lineplot(data=costs['payment'][0:62], label="expenses")
sns.lineplot(data=costs['value'][0:62], label="present")
x = [*range(0,88)]
sns.barplot(x = costs.index[0:62],
            y = costs['payment'][0:62],
            palette=["gray" if (int(x) >= 15 and int(x)<= 45) else 'w' for x in x])
plt.ylabel("Payment")
plt.xlabel("Time")
#sns.lineplot(data=costs)
```
